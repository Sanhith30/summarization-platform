```python
# Step 4: Create .env with GPU enabled
env_content = """
DEBUG=True
LOG_LEVEL=INFO
ENABLE_GPU=True
MODEL_CACHE_DIR=/kaggle/working/models
DATABASE_URL=sqlite:///./app.db
"""
with open('.env', 'w') as f:
    f.write(env_content)
print("Created .env with GPU enabled")
```

```python
# Step 5: Setup ngrok for public URL
from pyngrok import ngrok

# Get free ngrok token from https://ngrok.com (sign up free)
# ngrok.set_auth_token("YOUR_NGROK_TOKEN")

# Create tunnel
public_url = ngrok.connect(8000)
print(f"\nüåê Public URL: {public_url}")
print(f"üìö API Docs: {public_url}/docs")
```
